```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context="talk")
import graphviz

from sklearn.model_selection import train_test_split

from xgboost import XGBClassifier, plot_tree
```

# Import Datasets

```python
df = pd.read_csv("../Datasets/data_training.csv")
```

```python
X = df.drop(columns="label")
y = df["label"]
```

```python
clf = XGBClassifier(learning_rate= 0.03510000006,
        subsample= 0.86659999,
        n_estimators= 44,
        min_child_weight= 1,
        max_depth= 10,
        gamma= 3,
        colsample_bytree= 0.987899999998)
```

```python
model = clf.fit(X, y)
```

```python
plot_tree(model, num_trees=0)
fig = plt.gcf()
fig.set_size_inches(100, 100)
fig.savefig('tree.png')
```
