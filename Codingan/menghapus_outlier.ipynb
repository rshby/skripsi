## Mencoba Menghapus Outlier dan Normalisasi

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

pd.options.display.max_rows = 500
```

### Import Datasets

```python
df = pd.read_csv("../Datasets/heart.csv")
df
```

```python
df_tr = df.loc[:,["trestbps","age"]]
df_tr
```

## Menghapus Outlier pada kolom trestbps

```python
Q1 = df["trestbps"].quantile(0.25) #membuat Q1
Q3 = df["trestbps"].quantile(0.75) #membuat Q3

IQR = Q3 - Q1 #menghitung IQR
print(f"IQR = {IQR}\nQ1 = {Q1}\nQ3 = {Q3}")
```

```python
df_tr = df_tr[(df_tr["trestbps"]>=Q1-(1.5*IQR)) & (df_tr["trestbps"]<=Q3+(1.5*IQR))] #filter outlier
df_tr.reset_index(inplace=True) #reset index
df_tr.drop("index", axis=1, inplace=True) #hapus kolom index lama
```
