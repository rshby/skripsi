```python
data = pd.read_csv("../Datasets/heart.dat", sep=" ", header=None)
data.columns = df.columns
data = data.astype({
    "age": "int64",
    "sex": "int64",
    "cp": "int64",
    "trestbps": "int64",
    "chol": "int64",
    "fbs": "int64",
    "restecg": "int64",
    "thalach": "int64",
    "exang": "int64",
    "oldpeak": "float64",
    "slope": "int64",
    "ca": "int64",
    "thal": "int64",
    "target": "int64"
})

data["cp"] = data["cp"]-1
data["slope"] = data["slope"] -1
data["target"] = data["target"]-1
data["thal"] = data["thal"].replace({
    3: 1,
    6: 2,
    7: 3
})
```
